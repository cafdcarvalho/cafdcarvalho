```python
localizar = "Documentos/jupyter/testmapas.csv"
```

```python
local = pd.read_csv(localizar)
```

```python
local
```

```python
local = local[["LAT","LON","INF","MANE VALUE"]]
```

```python
map = folium.Map(location=[-22.8983135,-43.5889753], zoom_start=11)
```

```python
map
```

```python
colors = {
'Neg':'red',
'Pos':'blue',
'Med':'orange',    
}

map = folium.Map(location=[-22.8983135,-43.5889753,], zoom_start=11)


for _, local in local.iterrows():
    if local['INF'] in colors.keys():
        folium.Marker(
            location=[local['LAT'], local['LON']],
            popup=local['MANE VALUE'],
            icon=folium.Icon(color=colors[local['INF']])
        ).add_to(map)
```

```python
map.save('localmap.html')
```
